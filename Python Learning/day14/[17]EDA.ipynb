```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic['deck'].value_counts(dropna=False)
```

```python
titanic.isnull() # 결측치 확인
```

```python
# 누락데이터 개수 isnull.sum()
titanic.isnull().sum()
```

```python
#누락데이터 행, 열 삭제 dropna(axis, thresh = x)(누락 데이터가 아닌 값이 몇 개 미만일 경우에만 적용시키는 인수)
titanic_2 = titanic.dropna(axis=1, thresh=800)
titanic_2
```

```python
titanic.isnull().sum()
```

```python
#결측(누락데이터) 치환 df.fillna(대체값, inplace=True) _최대값, 최소값, 평균값 등등 대처
age_mean = titanic['age'].mean()
age_mean
```

```python
titanic['age'].fillna(age_mean, inplace=True)
```

```python
titanic.isnull().sum()
```

```python
titanic['embarked'].value_counts(dropna=False)
```

```python
titanic['embarked'].value_counts(dropna=False).idxmax()
```

```python
e_idxmax = titanic['embarked'].value_counts(dropna=False).idxmax()
titanic['embarked'].fillna(e_idxmax, inplace=True)
```

```python
#embark_town 도 최빈값으로 채워보자.
et_idxmax = titanic['embark_town'].value_counts(dropna=False).idxmax()
titanic['embark_town'].fillna(et_idxmax, inplace = True)
```
