```python
#여러 개의 선으로 표시  속성을 확인하자.  linestyle, color , marker
data = [0,1,4,9,16]
x = [0,1,2,3,4]
y = [0,1,2,3,4]

plt.plot(x, y, ls = '--', c = 'blue', marker = 'o', label = 'linear')
plt.plot(x, data, ls = ':', c = 'green', marker = '*', label = 'quad')
plt.legend()
plt.title("My First Plot")
plt.xlabel("X")
plt.ylabel('Y'),
plt.grid();
```

```python
# 곡선
x = np.arange(-np.pi, np.pi, 0.1)

plt.plot(x, np.cos(x), marker = 'o', label = "Cosin")
plt.plot(x, np.sin(x), marker = '*', label = "Sin")
plt.legend()
plt.title("Sin - Cos Plot")
plt.grid(True);
```

```python
# exam 01) emp.csv를 로드한 후 이름과 봉급을 깃점으로 선 그래프를 그린다.

import pandas as pd
emp = pd.read_csv('emp.csv')
#emp
emp_name = emp['ename']
emp_sal = emp['sal']

plt.plot(emp_sal, ls = '--', c = 'red', marker = 'o')
plt.bar(emp_name, emp_sal)
plt.xticks(rotation = 90)
plt.xlabel("Name"),
plt.ylabel("Salary")
plt.title("ENAME Salaries")

for i in range(len(emp_name)):
    height = emp_sal[i]
    plt.text(emp_name[i], height+0.5, '%.1f' %height, ha='center', va='bottom', size = 12)
    
plt.grid(True);
```

```python
# exam 02) 직업별로 그룹화 한 결과를 원 그래프로 그려서 수치를 범례 표시한다.

# 직업별 평균 연봉(sal * 12)

new_emp = emp
new_emp['Year_sal'] = emp['sal']*12
#new_emp[['job']]

result = new_emp.groupby('job')['Year_sal'].mean().sort_values(ascending = True).reset_index()
#result
#label = ['CLERK', 'SALESMAN', 'MANAGER', 'ANALYST', 'PRESIDENT']

plt.axis('equal')
plt.pie(result['Year_sal'], labels = result['job'], autopct = "%2.f%%")
plt.legend()
plt.title("EMP Year Salaries");
```

```python
# exam 03) 부서별 결과를 막대그래프로 표시한다. 단, 봉급의 합을 이용한다.

result = emp.groupby('deptno')['sal'].sum().reset_index()  # 부서별 sal 합계
print(result)

#deptno = emp['deptno'].unique()
#deptno

plt.bar(result['deptno'].astype('str'), result['sal'], tick_label = result['deptno'])
plt.xlabel("deptno")
plt.ylabel("Salary Sum")
plt.title("DEPT Salaries Sum")

for i in range(len(result['deptno'].astype('str'))):
    height = result['sal'][i]
    plt.text(result['deptno'].astype('str')[i], height+0.5, '%.1f' %height, ha='center', va='bottom', size = 12)
    
plt.grid(True);
```

```python
# 직업별 봉급의 최댓값
result = emp.groupby('job')['sal'].max().reset_index()
print(result)
plt.subplot(1,2,1)
plt.plot(result['job'].astype('category'), result['sal'], ls = '--', marker = 'o', c = 'r')
plt.xticks(rotation = 90)
plt.xlabel("job")
plt.ylabel("MAX Salary")
plt.title("Job Max Salaries")

for i in range(len(result['job'].astype('category'))):
    height = result['sal'][i]
    plt.text(result['job'].astype('category')[i], height+0.5, '%.1f' %height, ha='center', va='bottom', size = 12)
    
plt.grid(True);

# 부서별 봉급의 최댓값
result2 = emp.groupby('deptno')['sal'].max().reset_index()
print(result2)

plt.subplot(1,2,2)
plt.plot(result2['deptno'].astype('str'), result2['sal'], ls = '-', marker = '^')
plt.xlabel("Dept")
plt.ylabel("MAX Salary")
plt.title("Job Max Salaries")

for i in range(len(result2['deptno'].astype('str'))):
    height = result2['sal'][i]
    plt.text(result2['deptno'].astype('str')[i], height+0.5, '%.1f' %height, ha='center', va='bottom', size = 12)

plt.grid(True);
```
