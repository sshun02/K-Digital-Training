```python
from sklearn import preprocessing #머신러닝에서 표준화를 해주는 모듈
                                  #-/ 평균, 분산, 최대, 최소값
import numpy as np
```

<< 1. section scaler >> 

```python
X_train = np.array([[ 1., -1.,  2.],
                     [ 2.,  0.,  0.],
                     [ 0.,  1., -1.]])
```

```python
scaler = preprocessing.StandardScaler().fit(X_train) # 학습 실행하겠다.
scaler
```

```python
X_train
```

```python
scaler.mean_  # 열마다 평균값을 나타냄
```

```python
scaler.scale_ # 각 열의 값을 평균 0, 분산 1, 표준화를 시켜준다. 0~255 -> 0~1로 축소 
```

```python
res = preprocessing.scale(X_train)
print(res.max(axis=0))
print(res.min(axis=0))
print(res.mean(axis=0))
print(res.var(axis=0))
```

```python
<< 2. transform >>
```

```python
X_scaled = scaler.transform(X_train)
```

```python
X_scaled
```

```python
min_max_scaler = preprocessing.MinMaxScaler() # 최대값 1, 최소값 0으로 표준화
res = min_max_scaler.fit_transform(X_train)
print(res.max(axis=0))
print(res.min(axis=0))
print(res.mean(axis=0))
print(res.var(axis=0))
```

```python
max_abs_scaler = preprocessing.MaxAbsScaler()  # 최대값,최소값을 절대값1로 표준화
res = max_abs_scaler.fit_transform(X_train)    # 모든 요소의 이 양수
print(res.max(axis=0))
print(res.min(axis=0))
print(res.mean(axis=0))
print(res.var(axis=0))
```
