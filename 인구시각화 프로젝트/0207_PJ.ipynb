```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline

man_total = './pop/manPop_2022.csv'
man_total_data=pd.read_csv(man_total)
tenmillion_df= man_total_data[man_total_data['2022년12월_남자 인구수'] >= 1000000][['행정구역', '2022년12월_남자 인구수']]
tenmillion_df 

plt.rc('font', family='Apple SD Gothic Neo')
plt.bar(tenmillion_df['행정구역'], tenmillion_df['2022년12월_남자 인구수'], color='blue', alpha=0.5)

```

```python
months = ['1월', '2월', '3월', '4월', '5월', '6월', '7월', '8월', '9월', '10월', '11월', '12월']
man_total_data_list = []
for month in months:
    column_name = f'2022년{month}_남자 인구수'
    man_total_data_list.append(man_total_data[column_name].sum())
man_total_data_list
```

```python
plt.plot(months, man_total_data_list, color = 'orange', marker = '*')
plt.ticklabel_format(axis='y',useOffset=False, style='plain')
plt.title('2022년 남자 총 인구수 변동')
```

```python
ratio = tenmillion_df['2022년12월_남자 인구수']
labels = tenmillion_df['행정구역']

plt.pie(ratio, labels=labels, autopct='%.1f%%')
plt.show()
ratio
```

```python
import pandas as pd

total = './pop/totalPop_2022.csv'
total_data=pd.read_csv(total)
total_data.head()
```

```python
#2022 1월부터 12월 까지 지역별 인구수 비율 변화.
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline


t_tenmillion_df_1= total_data[total_data['2022년1월_총인구수'] >= 2000000][['행정구역', '2022년1월_총인구수']]

t_tenmillion_df_12= total_data[total_data['2022년12월_총인구수'] >= 2000000][['행정구역', '2022년12월_총인구수']]

plt.rc('font', family='Apple SD Gothic Neo')
plt.bar(t_tenmillion_df_1['행정구역'], t_tenmillion_df_1['2022년1월_총인구수'], color='blue', alpha=0.5)

```

```python
months = ['1월', '2월', '3월', '4월', '5월', '6월', '7월', '8월', '9월', '10월', '11월', '12월']
total_data_list = []
for month in months:
    column_name = f'2022년{month}_총인구수'
    total_data_list.append(total_data[column_name].sum())

plt.plot(months, total_data_list, color = 'orange', marker = '*')
plt.ticklabel_format(axis='y',useOffset=False, style='plain')
plt.title('2022 전체 인구수 변동')
```

```python
t_ratio_1= t_tenmillion_df_1['2022년1월_총인구수']
t_labels_1 = t_tenmillion_df_1['행정구역']

plt.pie(t_ratio_1, labels=t_labels_1, autopct='%.1f%%')
plt.show()

#인구 200만 이상인 지역을 기준으로 비율표시.
```

```python
t_ratio_12 = t_tenmillion_df_12['2022년12월_총인구수']
t_labels_12 = t_tenmillion_df_12['행정구역']

plt.pie(t_ratio_12, labels=t_labels_12, autopct='%.1f%%')
plt.show()
```

```python
fig, axes = plt.subplots(1,2)

plt.title('지역별 1월~12월 인구 분포도 변화')
axes[0].pie(t_ratio_1, labels=t_labels_1, autopct='%.1f%%')
axes[1].pie(t_ratio_12, labels=t_labels_12, autopct='%.1f%%')

plt.show()

# 별 차이 못느끼겠다.
```
