#### 남자 인구수

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

man_pop1 = pd.read_csv('./pop/manPop_2022.csv', index_col = '행정구역')  # 2019년 남자 시도별 인구 수
man_pop1
```

```python
# 2020년 남자 총 인구 수 변화
man_total = []
for month in range(1, 13):
    man_total.append(man_pop1['2022년{}월_남자 인구수'.format(month)].sum())
print(man_total)
```

```python
plt.rc('font', family = 'Apple SD Gothic Neo')
#plt.figure(figsize = (10,5))
plt.plot([str(i)+'월' for i in range(1, 13)], man_total, c = 'dodgerblue', marker = '*')
plt.grid(True, color='black', alpha=0.2, linestyle='--')
plt.ticklabel_format(axis='y',useOffset=False, style='plain')  # y축 지수표현 변경
plt.title('2022년 남자 총 인구수 변동')
plt.xlabel('월')
plt.ylabel('인구수(명)')
plt.show();
```

```python
plt.grid(True, color='black', alpha=0.2, linestyle='--')
plt.bar(np.arange(17), man_pop1['2022년1월_남자 인구수'], width = 0.3, color = 'plum', label = '1월')
plt.bar(np.arange(17)+0.3, man_pop1['2022년12월_남자 인구수'], width = 0.3, color = 'powderblue', label = '12월')
plt.xlabel('시도')
plt.ylabel('인구수')
plt.ticklabel_format(axis='y',useOffset=False, style='plain')  # y축 지수표현 변경
plt.xticks(np.arange(17)+0.3, man_pop1.index, rotation = 90)
plt.title("2022년 시도별 남자 인구 현황 비교(1월, 12월)")
plt.legend()
plt.show();
```

```python
fig, axes = plt.subplots(3, 4, constrained_layout = True)
fig.set_size_inches(20,18)

for row in range(3):
    for col in range(4):
        idx = row * 4 + col
        
        ax = axes[row][col]
        ax.grid(True)
        ax.set_title('2022년{}월 시도별 남자 인구'.format(idx+1))
        ax.bar(range(17), man_pop1['2022년{}월_남자 인구수'.format(idx+1)], color = 'plum')
        ax.set_xticks(np.arange(17), man_pop1.index, rotation = 90)
        ax.ticklabel_format(axis='y',useOffset=False, style='plain')  # y축 지수표현 변경
        ax.set_xlabel('시도')
        ax.set_ylabel('인구수');
```

#### 남자 / 여자 인구수 비교

```python
print("월별 남자 총 인구 : {}".format(man_total))
print("월별 여자 총 인구 : {}".format(woman_total))
```

```python
man_woman = pd.DataFrame([man_total, woman_total], columns = [str(i)+'월' for i in range(1, 13)],
                        index = ['남', '여'])
man_woman.index.name = '성별'
man_woman.sum(axis = 0)
#man_woman
```

```python
plt.title('2022년 1월 남녀 인구수 비율')
plt.pie(man_woman['1월'], autopct = '%.3f%%', colors = ['dodgerblue', 'magenta'],
       wedgeprops = dict(width=0.7, edgecolor='w'), textprops={'fontsize':15})
plt.legend(labels = man_woman.index)
plt.show();
```

```python
fig, axs = plt.subplots(3, 4, constrained_layout = True)
fig.set_size_inches(10, 9)

for row in range(3):
    for col in range(4):
        idx = 4 * row + col
        
        ax = axs[row][col]
        ax.pie(man_woman['{}월'.format(idx+1)], autopct = '%.3f%%', colors = ['dodgerblue', 'magenta'],
              wedgeprops = dict(width=0.7, edgecolor='w'), textprops={'fontsize':15})
        ax.set_title("2022년{}월 남녀 비율".format(idx+1))
        ax.legend(labels = man_woman.index);
```
