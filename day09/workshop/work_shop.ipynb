````python
# 출생아수
import os

for i in range(1880, 2016):
    filename = os.getcwd()+"\EXAM01\yob{}.txt".format(i)
    with open(filename, "r") as f:
        file = f.readlines()
        print("{0}년도 신생아 : {1}명".format(i, len(list(file))))
    
````

````python
# 성별 출생아수
for year in range(1880, 2017) :
    filename = os.getcwd()+"\EXAM01\yob{}.txt".format(year)
    with open(filename, "r") as f:
        babys = f.readlines()
        boys = [baby for baby in babys if baby.split(',')[1]=='M']
        girls = [baby for baby in babys if baby.split(',')[1]=='F']

        print("{0}년도 신생아 : 남아 {1}명, 여아 {2}명".format(year, len(boys), len(girls)))
````
